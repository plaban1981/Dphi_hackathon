## Data Description
The dataset is composed of 15 attributes in total, each inclusive of the eye state classified for each data point acquired using a video recording that was simultaneously done while collecting the EEG data using the EEG device. 

The attributes that the specified dataset contains are the channels in the EEG device used to record the data. Instances were input chronologically, enabling the analysis of useful test set for syntactic and semantic theories. 

There is a total of 14 electrodes attached, each labelled and placed on the following scalp areas: AF3, F7, F3, FC5, T7, P7, FC6, F4, F8, O1, O2, P8, T8 and AF4.
